```python
events = json.load(open("event_clean_train.json",'r'))
```

```python
first_dict = {}
second_dict = {}
third_dict = {}
fourth_dict = {}
four_tuple_dict ={}
```

```python
def dict_function(input_dict, token):
    if token in input_dict:
        input_dict[token] = input_dict[token]+1
    else:
        input_dict[token] = 1
```

```python
for i, event in enumerate(events):
    event_list = events[i][0]['event'].split(' ')
    four_tuple = events[i][0]['event']
    dict_function(first_dict, event_list[0])
    dict_function(second_dict, event_list[1])
    dict_function(third_dict, event_list[2])
    dict_function(fourth_dict, event_list[3])
    dict_function(four_tuple_dict, four_tuple)

```
